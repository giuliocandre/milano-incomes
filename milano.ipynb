```python
import pandas as pd

# Load the DataFrame from the Feather file
df_milano = pd.read_feather('milano.feather')

# Display the first few rows of the DataFrame
df_milano.head()
```

```python
df_milano.reset_index(drop=True, inplace=True)
columns = df_milano.columns
new_cols = columns[1:]
df_milano.drop(columns=[' '], inplace=True)
df_milano.columns = new_cols
df_milano.drop(columns=['Codice catastale', 'Codice Istat Comune', 'Denominazione Comune', 'Sigla Provincia', 'Regione', 'Codice Istat Regione'], inplace=True)
df_milano.drop(columns=[' '], inplace=True)

df_milano.head()
```

```python
df_milano_sum = df_milano.sum().to_frame().T
df_milano_sum.drop(columns=['CAP'], inplace=True)
df_milano_sum.head()
```

```python
df_milano_sum.columns
```

```python

buckets = ['Reddito complessivo minore o uguale a zero euro - Frequenza',
    'Reddito complessivo da 0 a 10000 euro - Frequenza',
    'Reddito complessivo da 10000 a 15000 euro - Frequenza',
    'Reddito complessivo da 15000 a 26000 euro - Frequenza',
    'Reddito complessivo da 26000 a 55000 euro - Frequenza',
    'Reddito complessivo da 55000 a 75000 euro - Frequenza',
    'Reddito complessivo da 75000 a 120000 euro - Frequenza',
    'Reddito complessivo oltre 120000 euro - Frequenza']

bucket_names = ['0', '1-10k', '10-15k', '15-26k', '26-55k', '55-75k', '75-120k', '120k+']
tot_contribuenti = df_milano_sum['Numero contribuenti']

import matplotlib.pyplot as plt
# Normalize the values by dividing by tot_contribuenti and multiplying by 100 to get percentages
df_milano_sum_normalized = df_milano_sum[buckets].div(tot_contribuenti.values[0]) * 100

# Plotting the histogram with bucket names as labels
df_milano_sum_normalized.T.plot(kind='bar', legend=False)
plt.xlabel('Income Buckets')
plt.ylabel('Normalized Values')
plt.title('Normalized Histogram of df_milano_sum for Income Buckets')
plt.xticks(ticks=range(len(bucket_names)), labels=bucket_names, rotation=45)
plt.show()
```

```python
stats = df_milano_sum_normalized.T
stats.columns = ['Percentage']
stats['Cumulative Percentage'] = stats['Percentage'].cumsum()
```

```python
from collections import defaultdict

cap_map = defaultdict(list)

for _, row in df.iterrows():
    for cap in row['CAP']:
        cap_map[cap].append((row['Vendita'], row['Affitto']))

average_map = {}
for cap, values in cap_map.items():
    avg_vendita = round(sum(v[0] for v in values)*1000 / len(values), 2)
    avg_affitto = round(sum(v[1] for v in values) / len(values), 2)
    average_map[cap] = (avg_vendita, avg_affitto)

```

```python
def get_avg(cap, rent=0):
    x = average_map.get(cap)
    if not x:
        return None
    return x[rent]


```

```python
df_milano['average_sale'] = df_milano['CAP'].apply(lambda cap: get_avg(cap, 0))
df_milano['average_rent'] = df_milano['CAP'].apply(lambda cap: get_avg(cap, 1))
df_milano['sale_to_income'] = df_milano['average_sale']/df_milano['Reddito Medio']*60
df_milano['sale_to_income'] = df_milano['sale_to_income'].apply(lambda x: round(x, 1))
df_milano['rent_yield'] = 100*df_milano['average_rent']*12/df_milano['average_sale']
df_milano['rent_yield'] = df_milano['rent_yield'].apply(lambda x: round(x, 2))
df_milano['Reddito Medio'] = df_milano['Reddito Medio']/1000
df_milano['Reddito Medio'] = 1000*df_milano['Reddito Medio'].apply(lambda x: round(x, 2))
df_milano[['CAP', 'sale_to_income']].to_csv('sale_to_income.csv', index=False)
df_milano['Medio/Mediano'] = df_milano['Reddito Medio']/df_milano['Reddito Mediano']
df_milano['Reddito Medio'] = df_milano['Reddito Medio'].apply(lambda x: int(x))
df_milano['Reddito Mediano'] = df_milano['Reddito Mediano'].apply(lambda x: int(x))
df_milano['sale_to_median'] = df_milano['average_sale']/df_milano['Reddito Mediano']*60
df_milano['sale_to_median'] = df_milano['sale_to_median'].apply(lambda x: round(x, 1))
```
